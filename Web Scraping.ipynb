# Disclaimer!!
* The work done above is not professional in any manner, since there are indexes under indexes if we extract the website by doing one index at a time it would be much time consuming and confusing. So if we do the whole ranking index under a single loop it will be more efficient!

## Making a Loop for entire "th" and its Components!
* I am doing it by two different methods.

```python
trs
```

#### Method # 1 (Loop)
* Since its a complicated loop, therefore I have mentioned its working in "#".

```python
for r in ranking:      # This is for all the ranking! Also this loop is to extract the "tr"!
    trs = r.find_all('tr')     # Assigning a variable for the "tr"!
    
    for th in trs[0].find_all('th'):   # This loop is for "th", since the "th" is only present in the first index in "tr"!
        print('{:20}'.format(th.text), end = ' ')   # This is the custom format to line-up the text! (20 is the character area)

    print()    # Move to next line!

    for tr in trs[1:]:                # As we are already done with the "0" index, therefore we are going to skip that one!
        
        for td in tr.find_all('td'):  # This one is for "td"!
            print('{:20}'.format(td.text), end = ' ')
        print()   # Move to next line!
        
    print('\n\n\n\n')     # This print is for the "main loop" a.k.a "first loop". It will create a gap b/w the "ranking chart"
```

#### Method # 2 (By Function)
* The functions is also a littel bit complicated but it is a much smarter way! And it also consist of loops almost identical to the loop we have deal above!

```python
def p_ranking(trs):                     # Making a function for "trs".
    for th in trs[0].find_all('th'):
        print('{:20}'.format(th.text), end = ' ')
    print()
    
    for tr in trs[1:]:
        for td in tr.find_all('td'):
            print('{:20}'.format(td.text), end = ' ')
        print()
```

```python
for r in ranking:                       # Making a loop for ranking
    print(r.caption.text)               # For Caption(Date)
    p_ranking(r.find_all('tr'))         # Giving the value of "tr" to the function!
    print('\n\n\n\n')
```

## For The Headings!
* In this step we are going to locate the division(div) of "class = ciPhotoContainer" and extract all the "h3"(headings) present in the "div"
* Following are the Headings

```python
format = html_soup.find_all('div' ,class_ = 'ciPhotoContainer')[0].find_all('h3')
format
```

#### Implimenting Heading in the Function(Method # 2)

```python
i = 0            # Assigning the value for the heading!
for r in ranking:
    print(r.caption.text)
    print(format[i].text)     # Now everytime when the loop starts it will print the heading of the following matches!
    p_ranking(r.find_all('tr'))
    print('\n\n\n\n')
    i+=1                 # Everytime the loop ends it will atomatically assaign a new value(number) for it (in order wise)!
```

## Converting The Scrap Data Into "CSV" Format!

```python
def csv_ranking(trs):                    # Making a new function for this!
    data = ""
    for th in trs[0].find_all('th'):
        data += th.text + ','           # Separating the values by ","!
    data = data[:-1]                    # Slicing the last variable from all the rows because of commas(,)!
    data += '\n'                        # Go to the next line!
    
    for tr in trs[1:]:
        for td in tr.find_all('td'):
            data += td.text + ','
        data = data[:-1]
        data += '\n'
    return data
```

```python
r = ranking[0]
data = csv_ranking(r.find_all('tr'))
print(data)
```

## To Save The Scrap In "CSV" Format!

* For single index

```python
file = format[0].text
with open(file+'.csv' , "w") as f:
    f.writelines(data)
```

* To save the whole "ranking" which contains multiple indexse we have to make a loop!

```python
for i,r in enumerate(ranking):
    file = format[i].text
    data = csv_ranking(r.find_all('tr'))
    with open(file+'.csv' , "w") as f:
        f.writelines(data)
```
